# Clean & Analyze Social Media

## Import required libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

## Generate random data for the social media data

```python
# define a list of categories

categories = ['Food', 'Travel', 'Music', 'Sports', 'News', 'Lifestyle', 'Technology', 'Science', 'Fashion', 'Health']

data = {
    'Date': pd.date_range('2021-01-01', periods=500),
    'Category': [random.choice(categories) for _ in range(500)],
    'Likes': np.random.randint(0, 10000, size=500)}
```

## Load the data into a Pandas DataFrame and Explore the data

```python
# create a Dataset
df = pd.DataFrame(data)


# display the first 5 rows of the DataFrame
df.head()
```

## Clean the data

```python
# reomve all null values
df = df.dropna()

# remove all dubplicates
df = df.drop_duplicates()

# convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# convert the dataframe likes to an integer
df['Likes'] = df['Likes'].astype(int)
```

## Visualize and Analyze the data

```python
# Create a histogram plot of the 'Likes' column
sns.histplot(df['Likes'], kde=True, color='green', binwidth=500)

# Add labels and title
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.title('Histogram of Likes')

# Add grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Create the boxplot
ax = sns.boxplot(x='Category', y='Likes', data=df)

# Set the title and labels for better understanding
ax.set_title('Boxplot of Likes by Category')
ax.set_xlabel('Category')
ax.set_ylabel('Likes')

# Rotate x-axis labels for better visibility
plt.xticks(rotation = 45)

# Display the plot
plt.show()
```

```python
# print mean of the Likes
print(df['Likes'].mean())

# print mean of the Likes per category
print(df.groupby('Category')['Likes'].mean())
```


##### Histogram of Likes Analysis:

* The histogram represents the frequency of a certain range of likes received.
* The x-axis represents the number of likes, ranging from 0 to 10,000.
* The y-axis represents the frequency of each like count.
* The green bars represent different ranges of like counts and their corresponding frequencies.
* There are peaks around approximately 2000, 6000 and slightly above 8000 likes indicating higher frequencies at these points.
* A smooth green curve is overlaid on top of the bars indicating a trend in data distribution.




##### Boxplot of Likes by Category Analysis:

* The graph represents the distribution of likes across different categories: Fashion, Technology, Travel, Sports, Music, Science, Health, Lifestyle, Food, and News.
* The y-axis represents the number of likes, ranging from 0 to 10,000.
* Each category has a blue box representing the interquartile range (IQR) of likes, with a line inside indicating the median value.
* There are no outliers indicated on any of the boxplots.
* The Travel category has the highest median value, indicating that posts in this category tend to receive more likes than posts in other categories.
* On the other hand, the News category has one of the lowest median values, suggesting that posts in this category generally receive fewer likes.


##### Cocnclusion

Based on the analysis of the two charts:

* **Boxplot of Likes by Category**: This chart shows that the Travel category tends to receive the most likes, while the News category generally receives fewer likes. This suggests that users are more engaged with travel-related content than news content.

* **Histogram of Likes**: This chart shows that there are certain ranges of likes (around 2000, 6000, and slightly above 8000) that are more common than others. This suggests that posts typically receive likes within these ranges.

In conclusion, the data suggests that travel-related content is more popular among users, as it tends to receive more likes. Additionally, the majority of posts receive likes within specific ranges, indicating that there might be common factors influencing user engagement. However, without additional context or data, such as the number of posts in each category or the type of posts (e.g., photos, videos, text updates), these conclusions are preliminary and further investigation would be needed to draw more definitive insights. Its also important to note that the number of likes is just one measure of engagement, and other factors such as comments and shares could also be significant.
